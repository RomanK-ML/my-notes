## Линейные графики

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2, 100)
f1 = x*2
f2 = x**2

fig, l_axes = plt.subplots(figsize=(14,6))
l_axes.set_title("Линейный график")   # Задать заголовок графика
l_axes.set_ylabel("y_label")          # Задать зоголовок для оси y
l_axes.set_xlabel("x_label")          # Задать зоголовок для оси x
l_axes.set_xlim(0.5, 3)               # Задать диапазон отображения по оси x
l_axes.set_ylim(-1, 10)               # Задать диапазон отображения по оси y
l_axes.plot(x, f1, color="blue", linewidth=2, linestyle=":", label="Пунктирная синяя линия")            # пунктирная синяя линия
l_axes.plot(x, f2, color="red", linewidth=2, linestyle="--", label="Штрихованная красная линия")        # штрихованная красная линия
l_axes.plot(x, f1+f2, color="black", linewidth=2, linestyle="-", label="Черная непрерывная линия")      # черная непрерывная линия
l_axes.grid()                         # Включить сетку на графике
l_axes.legend(title="Линии")          # Включить наименование графиков
plt.show()
```

https://seaborn.pydata.org/generated/seaborn.distplot.html

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

x = np.linspace(0, 2, 100)
f1 = x*2
f2 = x**2

l_axes = plt.figure(figsize=(14,10))

l_axes.title("Линейный график")

sns.displot(x, f1, color="blue", label="Пунктирная синяя линия")
plt.ylabel("y_label")
plt.xlabel("x_label")
plt.legend(title="Линии")
plt.show()
```

## Диаграмма с областями

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

year = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2018]
population_by_continent = {
    'africa': [228, 284, 365, 477, 631, 814, 1044, 1275],
    'americas': [340, 425, 519, 619, 727, 840, 943, 1006],
    'asia': [1394, 1686, 2120, 2625, 3202, 3714, 4169, 4560],
    'europe': [220, 253, 276, 295, 310, 303, 294, 293],
    'oceania': [12, 15, 19, 22, 26, 31, 36, 39],
}

fig, stack_axes = plt.subplots(figsize=(14,6))
stack_axes.set_title("Диаграмма с областями")   # Задать заголовок графика
stack_axes.set_ylabel("population")          # Задать зоголовок для оси y
stack_axes.set_xlabel("years")          # Задать зоголовок для оси x
#stack_axes.set_xlim(0.5, 3)               # Задать диапазон отображения по оси x
#stack_axes.set_ylim(-1, 10)               # Задать диапазон отображения по оси y
stack_axes.stackplot(year, population_by_continent.values(), labels=population_by_continent.keys(), alpha=0.8)
stack_axes.grid()                         # Включить сетку на графике
stack_axes.legend()                       # Включить наименование графиков
plt.show()
```

## Диаграмма рассеяния

## Столбчатые диаграммы

```python
import numpy as np
import matplotlib.pyplot as plt

labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels))
width = 0.35


fig, bar_axes = plt.subplots(figsize=(14,6))
bar_axes.set_title("Столбчатые диаграммы")   # Задать заголовок графика
bar_axes.set_ylabel("Scores")          # Задать зоголовок для оси y
bar_axes.set_xlabel("x_label")          # Задать зоголовок для оси x
#bar_axes.set_xlim(0.5, 3)               # Задать диапазон отображения по оси x
bar_axes.set_ylim(0, 40)               # Задать диапазон отображения по оси y
bar1 = bar_axes.bar(x - width/2, men_means, width, label="Men")
bar2 = bar_axes.bar(x + width/2, women_means, width, label="Women")
bar_axes.set_xticks(x, labels)
bar_axes.bar_label(bar1, padding=3)
bar_axes.bar_label(bar2, padding=3)
bar_axes.legend()          # Включить наименование графиков
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

fig, bar_axes = plt.subplots(figsize=(14,6))
bar_axes.set_title("Столбчатые диаграммы")   # Задать заголовок графика
bar_axes.set_ylabel("Scores")          # Задать зоголовок для оси y
bar_axes.set_xlabel("x_label")          # Задать зоголовок для оси x
#bar_axes.set_xlim(0.5, 3)               # Задать диапазон отображения по оси x
#bar_axes.set_ylim(0, 40)               # Задать диапазон отображения по оси y
bar_axes.barh(y_pos, performance, xerr=error, align='center')
bar_axes.set_yticks(y_pos, labels=people)
bar_axes.invert_yaxis()  # labels read top-to-bottom
plt.show()
```

## Контурные диаграммы

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats


means = (0.5, 0.9)
covar = [
    [1., 0.6],
    [0.6, 1.]
]

mvn = stats.multivariate_normal(means, covar)
x, y = np.meshgrid(
    np.linspace(-3, 3, 80),
    np.linspace(-2, 4, 80)
)
data = np.dstack((x, y))

fig, contour_axes = plt.subplots(figsize=(14,6))
contour_axes.set_title("Контурная диаграмма")   # Задать заголовок графика
contour_axes.set_ylabel("y_label")          # Задать зоголовок для оси y
contour_axes.set_xlabel("x_label")          # Задать зоголовок для оси x
contour_axes.set_xlim(-4, 4)               # Задать диапазон отображения по оси x
contour_axes.set_ylim(-2, 4)               # Задать диапазон отображения по оси y
contour_axes.contour(x, y, mvn.pdf(data), levels=10)
contour_axes.grid()                         # Включить сетку на графике
plt.show()
```

## Круговая диаграмма

```python
import numpy as np
import matplotlib.pyplot as plt

names = 'Tom', 'Dick', 'Harry', 'Jill', 'Meredith', 'George'
speed = [8, 7, 12, 4, 3, 2]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'red', 'orange']
explode = (0.1, 0, 0, 0, 0, 0)

fig, pie_axes = plt.subplots(figsize=(14,6))
pie_axes.set_title("Круговая диограмма")# Задать заголовок графика
pie_axes.set_xlim(0, 5)                 # Задать диапазон отображения по оси x
pie_axes.set_ylim(-1, 6)                # Задать диапазон отображения по оси y
pie_axes.pie(speed, explode=explode, labels=names, colors=colors, autopct="%1.1f%%", shadow=True, startangle=140)
fig.set_facecolor('white')
pie_axes.grid()                         # Включить сетку на графике
pie_axes.legend(title="Title")          # Включить наименование графиков
plt.show()
```

## Тепловые карты

```python
import numpy as np
import numpy.random
import matplotlib.pyplot as plt


temperature = np.random.randn(4096)
anger = np.random.randn(4096)
heatmap, xedges, yedges = np.histogram2d(temperature, anger, bins=(64,64))
extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]

plt.clf()
plt.ylabel('Anger')
plt.xlabel('Temp')
plt.imshow(heatmap, extent=extent)
plt.show()
```
